```python
import numpy as np
from sklearn.metrics import confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
model_folder = "../training/saved/2024-03-26/train_80_20_1_8_octaves"
archive = np.load(model_folder + "/predictions.npz")
y_pred = archive["y_pred"]
y_gt = archive["y_gt"]
```

```python
def plot_confusion_matrix(pred, gt, title, labels=None):
    if labels is None:
        labels = []
    table = pd.DataFrame(confusion_matrix(gt, pred))
    plt.figure(figsize=(20, 15))
    ax = sns.heatmap(table, annot=True, fmt='d', cmap='viridis',  annot_kws={"size": 20})
    ax.set_title(title)
    if len(labels) != 0:
        ax.set_xticklabels(labels)
        ax.set_yticklabels(labels, rotation=0)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.show()
```

Confusion matrix for E string

```python
E_pred = []
E_gt = []

for i in range(len(y_pred)):
    pred_index = np.argmax(y_pred[i][0])
    gt_index = np.argmax(y_gt[i][0])
    if gt_index > 0:
            E_pred.append(pred_index-1)
            E_gt.append(gt_index-1)

E_pred = np.array(E_pred)
E_gt = np.array(E_gt)
E_labels = [str(i) + 'E' for i in range(15)]
plot_confusion_matrix(E_pred, E_gt, 'E confusion matrix', labels=E_labels)
```
