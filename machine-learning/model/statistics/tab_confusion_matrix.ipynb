```python
test_cases = [0,1,2,3,4,5,'train_83p']
y_pred = np.empty((0,6,21))
y_gt = np.empty((0,6,21))

for tc in test_cases:
    print(f"start {tc}")
    model_folder = f"../training/saved/2024-04-03/{tc}_1_8_octaves"
    archive = np.load(model_folder + "/predictions.npz")
    print(f"{tc}: loaded archive")
    y_pred = np.concatenate((y_pred, archive["y_pred"]), axis=0)
    print(f"{tc}: loaded pred")
    y_gt = np.concatenate((y_gt, archive["y_gt"]), axis=0)
    print(f"{tc}: loaded gt")
```

Confusion matrix for E string

```python
E_pred = []
E_gt = []

for i in range(len(y_pred)):
    pred_index = np.argmax(y_pred[i][0])
    gt_index = np.argmax(y_gt[i][0])
    if gt_index > 0 or pred_index > 0:
            E_pred.append(pred_index-1)
            E_gt.append(gt_index-1)

E_pred = np.array(E_pred)
E_gt = np.array(E_gt)
E_labels = ['-1'] + [str(i) for i in range(0,14)]
plot_confusion_matrix(E_pred, E_gt, 'matrice confuzie coarda E2', labels=E_labels, xlabel='index fret prezis', ylabel='index fret real')
```
