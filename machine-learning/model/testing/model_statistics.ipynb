```python
import math
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import pandas as pd
import seaborn as sns
```

```python
model_folder = "../training/saved/2024-03-06/2_4"
archive = np.load(model_folder + "/predictions.npz")
y_pred = archive["y_pred"]
y_gt = archive["y_gt"]
```

Confusion matrix for E string

```python
E_pred = []
E_gt = []

for i in range(len(y_pred)):
    pred_index = np.argmax(y_pred[i][0])
    gt_index = np.argmax(y_gt[i][0])
    if gt_index > 0:
            E_pred.append(pred_index)
            E_gt.append(gt_index)

E_pred = np.array(E_pred)
E_gt = np.array(E_gt)

table = pd.DataFrame(confusion_matrix(E_gt, E_pred))
sns.heatmap(table, annot=True, fmt='d', cmap='viridis')
```
